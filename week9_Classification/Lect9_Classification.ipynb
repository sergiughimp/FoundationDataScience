```python
from sklearn.datasets import make_blobs as blobs
import matplotlib.pyplot as plt

# Generate synthetic data
X, y = blobs(n_samples=300, centers=4, n_features=2, random_state=0, cluster_std=1.0)

# Plot the clusters
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis', s=60, edgecolors='black')

plt.title("Clusters generated with make_blobs", fontsize=14)
plt.xlabel("X_1")
plt.ylabel("X_2")
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
def softmax(x_values):
    exp_values = [math.exp(x) for x in x_values]
    total = sum(exp_values)
    return [x / total for x in exp_values]

print(softmax([2.0, 1.0, 0.1]))
```

```python
# 1. Import libraries
from sklearn.datasets import make_blobs as blobs
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt
import numpy as np

# 2. Generate synthetic data (same as before)
X, y = blobs(n_samples=300, centers=4, n_features=2, random_state=0, cluster_std=1.0)

# 3. Create and train the Logistic Regression model
model = LogisticRegression()
model.fit(X, y)

# 4. Create a mesh grid for visualization
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02),
                     np.arange(y_min, y_max, 0.02))

# 5. Predict classes for each grid point
Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# 6. Plot decision boundaries and training data
plt.figure(figsize=(8, 6))
plt.contourf(xx, yy, Z, cmap='viridis', alpha=0.4)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis', edgecolors='black', s=60)
plt.title("Logistic Regression Decision Boundaries", fontsize=14)
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
from sklearn.datasets import make_blobs as blobs
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt
import numpy as np

# 1. Generate synthetic dataset
X, y = blobs(n_samples=300, centers=4, n_features=2, random_state=0, cluster_std=1.0)

# 2. Create and train the Decision Tree
classifier = DecisionTreeClassifier(max_depth=4)
classifier.fit(X, y)

# 3. Create a mesh grid for visualization
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02),
                     np.arange(y_min, y_max, 0.02))

# 4. Predict classes for each grid point
Z = classifier.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# 5. Plot the decision boundaries and data points
plt.figure(figsize=(8, 6))
plt.contourf(xx, yy, Z, cmap='viridis', alpha=0.4)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis', edgecolors='black', s=60)
plt.title("Decision Tree Classifier (max_depth=4)", fontsize=14)
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```
