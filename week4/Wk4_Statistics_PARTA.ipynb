## 2. Descriptive measures

Consider either an array `x` or a single-column data frame `y` containing 10 values:

```python
y = pd.DataFrame([87, 82, 89, 90, 94, 92, 86, 92, 80, 96], columns=['values'])
```

### 2.1. Mean & median

```python
mean = y['values'].mean()
print(f'The mean value is {mean}.')
```

The steps to find the ___median___ value of an array are:

1. Sort the array

2. If the array has an odd number of elements (e.g. 1, 3, 5, 7, 9, etc.) return the middle point. Otherwise, return the average of the two middle points.

```python
# Case 2a
data = np.array([1, 2, 3])
np.median(data)
```

```python
# Case 2b
data = np.array([2, 1, 3, 4])
np.median(data)
```

Back to our example:

```python
# Sorted values are (80, 82, 86, 87, 89, 90, 92, 92, 94, 96)
median = y['values'].median()
print(f'The median value is {median}')
```

__Note.__ You can use the `mode()` function on a data frame column to return the most frequest item.

```python
# Sorted values are (80, 82, 86, 87, 89, 90, 92, 92, 94, 96)
y['values'].mode()
```

### 2.2. Variance & Standard Deviation

Let's look into measures of variability aka ___dispersion___. For example, we can find the __range__ of our values:

```python
# Sorted values are (80, 82, 86, 87, 89, 90, 92, 92, 94, 96)
minimum = y['values'].min()
maximum = y['values'].max()

print(f"The range of values is {maximum-minimum} and the minimum and maximum are: [{minimum}, {maximum}]")
```

Let's compute the variance $\sigma^{2}$, and standard deviation $\sigma$, of our values:

* __Standard deviation__ measures the spread of numbers from the mean. It is the square root of variance. 


* __Variance__ is the average of the squared differences from the mean.

```python
v = y['values'].var()
s = y['values'].std()

print(f'The variance is {v:.2f}')
print(f'The standard deviation is {s:.2f}')
```

__Note.__ The variance is the standard deviation squared:

```python
# subtract the variance from the squared standard deviation:
# the result is close to zero i.e. identical entities
abs(v - s**2) < 1E-09
```

## 4. Order statistics

Given $n$ sample values and $0 < p < 1$, then the $100p$-th percentile has $np$ values less than it and $n(1-p)$ values greater than it.

* $p=0.25$ or 25-th pecentile or 1st quartile


* $p=0.5$ or 50-th percentile or 2nd quartile or median


* $p=0.75$ or 75-th percentile or 3rd quartile

```python
# Set seed
np.random.seed(123456789)

y = pd.DataFrame(np.random.normal(loc=2., scale=1., size=1000), columns=['values'])

assert all(y.quantile(0.5) == y.median())
y.quantile(0.25)

fig,ax = plt.subplots()

ax.set(xlim=[-1, 5], xticks=[-1, 0, 1, 2, 3, 4, 5])

# Plot 5-th, 25-th, 50-th, 75-th, 95-th percentiles
y.boxplot(ax=ax, 
          vert=False,
          grid=False,
          whis=(5, 95),
          showmeans=True,
          figsize=(1, 10))
```

## 5. Normal distribution

Let's create a normal distribution using the `np.random.normal()` function. The function takes three arguments:

* `loc  ` is the mean of the distribution;
* `scale` is the standard deviation; and
* `size ` is the number of elements to generate.

```python
# Set seed
np.random.seed(123456789)

# Populate a numpy array with 1,000 elements
data = np.random.normal(loc=0, scale=1, size=1000)

print(f'A distribution of {len(data)} elements:')
print(f'- mean is {data.mean():.2f}')
print(f'- standard deviation is {data.std():.2f}')
```

Let's plot a histogram of the data using `seaborn`'s `histplot()` (or, similarly, `distplot()`) function:

```python
fig, ax = plt.subplots()

ax.set(xlim=[-4, 4], xlabel='Value')
sns.histplot(data, kde=True, ax=ax);
```

### A.2. Uniform distribution

```python
from scipy.stats import uniform


samples = np.random.uniform(0,1,100)

assert np.all(samples >= 0)
assert np.all(samples <  1)

_, ax = plt.subplots()

ax.set(xlim=[-1, 2],
       ylim=[0, 2],
       xlabel='x',
       ylabel='Density')

ax.vlines([0, 1],
          [0, 0],
          [uniform.pdf(0), uniform.pdf(1)], 
          colors=['coral'],
          linestyles=['dashed', 'dashed'])

ax.plot(samples, uniform.pdf(samples));
```

### A.3. Normal distribution

```python
from scipy.stats import norm


xmin = -5.
xmax =  5.

x = np.linspace(xmin, xmax, 100)

y1 = norm.pdf(x, 0.0, 1.0)
y2 = norm.pdf(x, 0.0, 2.0)
y3 = norm.pdf(x, 2.0, 0.5)

_, ax = plt.subplots()

ax.set(xlim=[-5, 5], 
       ylim=[0, 1], 
       xlabel='x', 
       ylabel='Density');

ax.plot(x, y1, color='coral', linewidth=2.0, linestyle='-',  label="$\mu = 0, \sigma=1$")
ax.plot(x, y2, color='blue',  linewidth=2.0, linestyle='--', label="$\mu = 0, \sigma=2$")
ax.plot(x, y3, color='black', linewidth=2.0, linestyle=':',  label="$\mu = 2, \sigma={}^1/{}_2$")

ax.legend();
```

### A.4. $t$-student distribution

Some __notes__ about $t$-distributions:

* $v$ is the _degrees of freedom_. As $v$ increases ($v \rightarrow \inf$), the $t$-distribution becomes a Gaussian.

* The $t$-distribution can be interpreted as an infinite mixture of Gaussians with the same mean but different precision.

* The $t$-distribution has longer tails that a Gaussian, making it more __robust__: it is not sensitive to outliers.

```python
from scipy.stats import t

xmin = -5.
xmax =  5.

x = np.linspace(xmin, xmax, 50)

_, ax = plt.subplots()

ax.set(xlim=[-5, 5], 
       ylim=[0, 1], 
       xlabel='x', 
       ylabel='Density')

ax.plot(x, t.pdf(x,  0.1), color='black', linewidth=2.0, linestyle=':',  label="$v = 0.1$")
ax.plot(x, t.pdf(x,  1.0), color='coral', linewidth=2.0, linestyle='--', label="$v=1$")
ax.plot(x, t.pdf(x, 99.0), color='blue',  linewidth=2.0, linestyle='-',  label="$v=99$")

ax.legend();
```

### A.5. $\chi^2$ (chi-square) distribution

```python
from scipy.stats import chi2

xmin = 0.
xmax = 10.

x = np.linspace(xmin, xmax, 1000)

_, ax = plt.subplots()

ax.set(xlim=[0, 10.], 
       ylim=[0, 0.5], 
       xlabel='x', 
       ylabel='Density')

ax.plot(x, chi2.pdf(x,  1), color='black', linewidth=2.0, linestyle=':',  label="$k = 1$")
ax.plot(x, chi2.pdf(x,  2), color='red',   linewidth=2.0, linestyle='--', label="$k = 2$")
ax.plot(x, chi2.pdf(x,  3), color='blue',  linewidth=2.0, linestyle='-',  label="$k = 3$")
ax.plot(x, chi2.pdf(x,  4), color='green', linewidth=2.0, linestyle='-',  label="$k = 4$")
ax.plot(x, chi2.pdf(x,  9), color='coral', linewidth=2.0, linestyle='-',  label="$k = 9$");
```
