```python
import pandas as pd
df = pd.read_csv("Amazon_Reviews.csv")
df.head()
```

```python
import re
def clean_text(text):
    text = text.lower()
    text = re.sub(r'[^a-z0-9\s]', '', text)
    text.replace("  ", "")
    return text
```

```python
df['cleaned_text'] = df['Text'].apply(clean_text)
```

```python
df['cleaned_text'][1]
```

```python
#tf-idf

from sklearn.feature_extraction.text import TfidfVectorizer

# initialize vetoriser

vectoriser = TfidfVectorizer(stop_words='english', max_features=5000)
X  = vectoriser.fit_transform(['cleaned_text'])
```

```python
df_filtered = df[(df['Score']) >= 4 | (df['Score'] <= 2)]
# df_filtered.head()

X_filtered = vectoriser.fit_transform(df_filtered['cleaned_text'])
y = df_filtered['Score'].apply(lambda s: "Positive" if s >= 4 else "Negative")
```

```python
X_filtered.shape
```

```python
# data splitting
from sklearn.model_selection import train_test_split
# actual classifier
from sklearn.neighbors import KNeighborsClassifier
# metric
from sklearn.metrics import accuracy_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_filtered, y, test_size=0.2, random_state=42)
```

```python
knn = KNeighborsClassifier(n_neighbors=10)
knn.fit(X_train, y_train)
```

```python
# predictions
y_predict = knn.predict(X_test)
y_predict
```
