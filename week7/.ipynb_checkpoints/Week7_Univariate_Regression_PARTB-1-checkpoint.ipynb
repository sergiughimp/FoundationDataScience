# Univariate Regression: Linear and Polynomial Models

This notebook demonstrates univariate regression using linear and polynomial models with `statsmodels`. Comments are included to show equivalent `scikit-learn` code.

```python

import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
from sklearn.preprocessing import PolynomialFeatures
```

## Part 1: Simulated Data - Linear and Polynomial Regression

```python

# Generate Simulated Linear Data
np.random.seed(0)
x_min=0
x_max=10
N=50 # Number of datapoints

x_linear = np.linspace(x_min, x_max,N)
y_linear = 3 * x_linear + 5 + np.random.normal(0, 3, x_linear.shape)

# Fit Linear Regression
X_linear_sm = sm.add_constant(x_linear)
model_linear_sm = sm.OLS(y_linear, X_linear_sm).fit()

# Display Linear Regression Summary
print("Linear Regression Summary (Simulated Linear Data):")
print(model_linear_sm.summary())

# Equivalent using scikit-learn:
# from sklearn.linear_model import LinearRegression
# linear_model = LinearRegression().fit(x_linear[:, np.newaxis], y_linear)

```

```python
# Generate Simulated Non-linear Data
np.random.seed(0)

x_nonlinear = np.linspace(x_min, x_max,N)
y_nonlinear = np.sin(x_nonlinear) + np.random.normal(0, 0.5, x_nonlinear.shape)

# Fit Polynomial Regression
order=4
X_poly_sm = sm.add_constant(PolynomialFeatures(order).fit_transform(x_nonlinear[:, np.newaxis]))
model_poly_sm = sm.OLS(y_nonlinear, X_poly_sm).fit()

# Display Polynomial Regression Summary
print("\nPolynomial Regression Summary (Simulated Non-linear Data, Degree 4):")
print(model_poly_sm.summary())

# Equivalent using scikit-learn:
# from sklearn.pipeline import make_pipeline
# poly_model = make_pipeline(PolynomialFeatures(4), LinearRegression())
# poly_model.fit(x_nonlinear[:, np.newaxis], y_nonlinear)
```

### Predictions and Visualization for Simulated Data

```python
# Predictions using statsmodels
y_linear_pred_sm = model_linear_sm.predict(X_linear_sm)
x_nonlinear_poly_sm = sm.add_constant(PolynomialFeatures(order).fit_transform(x_nonlinear[:, np.newaxis]))
y_nonlinear_pred_sm = model_poly_sm.predict(x_nonlinear_poly_sm)

# Plot Simulated Data
plt.figure(figsize=(14, 6))

# Linear Data Plot
plt.subplot(1, 2, 1)
plt.scatter(x_linear, y_linear, color="blue", label="Data")
plt.plot(x_linear, y_linear_pred_sm, color="orange", label="Linear Fit")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Simulated Linear Relationship")
plt.legend()

# Non-linear Data Plot
plt.subplot(1, 2, 2)
plt.scatter(x_nonlinear, y_nonlinear, color="blue", label="Data")
plt.plot(x_nonlinear, y_nonlinear_pred_sm, color="orange", label="Polynomial Fit")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Simulated Non-linear Relationship")
plt.legend()

plt.tight_layout()
plt.show()
```

## Part 2: Real Data Analysis - Environmental Dataset

```python

import pandas as pd

# Load environmental dataset
data = pd.read_csv('Enviromental_dataset.csv')
data_cleaned = data.dropna(subset=['pm2.5', 'TEMP', 'PRES'])

# Prepare TEMP vs PM2.5 data
X_temp = data_cleaned['TEMP'].values
y_pm25 = data_cleaned['pm2.5'].values

# Prepare PRES vs TEMP data
X_pres = data_cleaned['PRES'].values
y_temp = data_cleaned['TEMP'].values
```

### Temperature (TEMP) vs PM2.5

```python

# Linear Regression (TEMP vs PM2.5)
X_temp_sm = sm.add_constant(X_temp)
model_temp_pm25_linear = sm.OLS(y_pm25, X_temp_sm).fit()
print("Linear Regression Summary (TEMP vs PM2.5):")
print(model_temp_pm25_linear.summary())
```

```python
# Polynomial Regression (Degree 3) (TEMP vs PM2.5)
X_temp_poly_sm = sm.add_constant(PolynomialFeatures(3).fit_transform(X_temp[:, np.newaxis]))
model_temp_pm25_poly = sm.OLS(y_pm25, X_temp_poly_sm).fit()
print("\nPolynomial Regression Summary (TEMP vs PM2.5, Degree 3):")
print(model_temp_pm25_poly.summary())

# Equivalent scikit-learn code:
# linear_model_temp_pm25 = LinearRegression().fit(X_temp[:, np.newaxis], y_pm25)
# poly_model_temp_pm25 = make_pipeline(PolynomialFeatures(3), LinearRegression())
# poly_model_temp_pm25.fit(X_temp[:, np.newaxis], y_pm25)
```

#### Predictions and Visualization for TEMP vs PM2.5

```python
# Predictions for TEMP vs PM2.5
y_pm25_pred_linear = model_temp_pm25_linear.predict(X_temp_sm)
y_pm25_pred_poly = model_temp_pm25_poly.predict(X_temp_poly_sm)

# Plot TEMP vs PM2.5
plt.figure(figsize=(10, 6))
plt.scatter(X_temp, y_pm25, color="blue", alpha=0.3, label="Observed Data")
plt.plot(X_temp, y_pm25_pred_linear, color="orange", label="Linear Fit")
plt.plot(X_temp, y_pm25_pred_poly, color="purple", label="Polynomial Fit (Degree 3)")
plt.xlabel("Temperature (TEMP)")
plt.ylabel("PM2.5 Concentration")
plt.title("Temperature vs PM2.5")
plt.legend()
plt.show()
```

### Pressure (PRES) vs Temperature (TEMP)

```python

# Linear Regression (PRES vs TEMP)
X_pres_sm = sm.add_constant(X_pres)
model_pres_temp_linear = sm.OLS(y_temp, X_pres_sm).fit()
print("Linear Regression Summary (PRES vs TEMP):")
print(model_pres_temp_linear.summary())
```

```python
# Polynomial Regression (Degree 3) (PRES vs TEMP)
X_pres_poly_sm = sm.add_constant(PolynomialFeatures(3).fit_transform(X_pres[:, np.newaxis]))
model_pres_temp_poly = sm.OLS(y_temp, X_pres_poly_sm).fit()
print("\nPolynomial Regression Summary (PRES vs TEMP, Degree 3):")
print(model_pres_temp_poly.summary())

# Equivalent scikit-learn code:
# linear_model_pres_temp = LinearRegression().fit(X_pres[:, np.newaxis], y_temp)
# poly_model_pres_temp = make_pipeline(PolynomialFeatures(3), LinearRegression())
# poly_model_pres_temp.fit(X_pres[:, np.newaxis], y_temp)
```

#### Predictions and Visualization for PRES vs TEMP

```python

# Predictions for PRES vs TEMP
y_temp_pred_linear = model_pres_temp_linear.predict(X_pres_sm)
y_temp_pred_poly = model_pres_temp_poly.predict(X_pres_poly_sm)

# Plot PRES vs TEMP
plt.figure(figsize=(10, 6))
plt.scatter(X_pres, y_temp, color="blue", alpha=0.3, label="Observed Data")
plt.plot(X_pres, y_temp_pred_linear, color="orange", label="Linear Fit")
plt.plot(X_pres, y_temp_pred_poly, color="purple", label="Polynomial Fit (Degree 3)")
plt.xlabel("Pressure (PRES)")
plt.ylabel("Temperature (TEMP)")
plt.title("Pressure vs Temperature")
plt.legend()
plt.show()
```
