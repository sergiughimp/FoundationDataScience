## The goal is to generalise, not overfit

The goal of machine learning is __good generalisation__; that is, the ability to predict new data.

## Preamble

```python
%matplotlib inline

import matplotlib.pyplot as plt
import seaborn as sns

import numpy as np

sns.set(style='ticks', font='Helvetica', font_scale=2)
```

## Linear regression

The simple __linear regression__ model involves a linear combination of input variables:

\begin{equation}
y(\mathbf{x}, \mathbf{w}) = w_0 + w_1x_1 + w_2x_2 + \dots + w_dx_d
\end{equation}

where $\mathbf{x} = (x_1, \dots, x_d)^T$ is a $d$-dimensional vector.

The key property is that $y(\mathbf{x}, \mathbf{w})$ is a __linear function__ of parameters ${w}_i$. But it is also a linear function of variables $x_i$, which is a limiting factor. 

So, we consider linear combinations of non-linear functions $\phi_i(\mathbf{x})$:

\begin{equation}
y(\mathbf{x}, \mathbf{w}) = w_0 + w_1\phi_1(\mathbf{x}) + w_2\phi_2(\mathbf{x}) + \dots + w_{m-1}\phi_{m-1}(\mathbf{x})
\end{equation}

where $\phi_i(\mathbf{x})$ are known as __basis functions__. The size of the model (that is, the number of parameters) is $m$.

For example, if $\phi_j(x) = x^j$ then:

\begin{equation}
y(\mathbf{x}, \mathbf{w}) = w_0 + w_1x + w_2x^2 + \dots + w_{m-1}x^{m-1}
\end{equation}

is a __polynomial__ regression model.

In a similar fashion, we can use a Gaussian basis function (see [Handbook](https://jakevdp.github.io/PythonDataScienceHandbook/05.06-linear-regression.html) for an example).

### A straight line

The simplest example is to fit a straight line to data:

\begin{equation}
y = w_0 + w_1x
\end{equation}

where $w_{0}$ is known as the _intercept_ and $w_1$ as the slope of the line.

Let's generate some data for $y(x) = -5 + 2x$: 

```python
def config(axis):
    """
    Simple plot configuration for y(x) = -5 + 2x
    """
    axis.set(ylabel='y(x)',
             xlabel='x',
             xticks=np.arange( 0, 12, 2),
             yticks=np.arange(-5, 20, 5))


# Create a random number generator (with given seed)
rng = np.random.RandomState(123456789)

# Generage 50 random values for x-axis (from 0 to 10)
x = 10 * rng.rand(50)

# Compute y(x), adding random noise to each result
y = 2 * x - 5 + rng.randn(50)

_, ax = plt.subplots()

# Configure the plot
config(ax)

# Plot the (x, y) data points
ax.scatter(x, y, color='#3B5B72');
```

Our __first, simple task__ is to approximate the line $y(x) = -5 + 2x$:

```python
from sklearn.linear_model import LinearRegression

# Create a linear regression model
model = LinearRegression(fit_intercept=True)

# Train it
model.fit(x[:, np.newaxis], y)

# Predict y-value for 1,000 data points in range from 0 to 10
u = np.linspace(0, 10, 1000)
v = model.predict(u[:, np.newaxis])

_, ax = plt.subplots()

# Configure the plot
config(ax)

# Overlay the predicted line over the input data
ax.scatter(x, y, color='#3B5B72')
ax.plot(u, v, color='#B7BF35', linewidth=3);
```

The __predicted model__, depicted above with the green line, is:

```python
print(f'y(x) = {model.intercept_:.3f} + {model.coef_[0]:.3f}x')
```

Recall that we wanted to predict $y(x) = -5 + 2x$. The is pretty close then. How does it work?

### Polynomial Curve fitting

Let's consider the function $y(x) = \sin(2\pi x)$. 

We generate 10 values for $x$ and the corresponding $y(x)$, adding random normal noise to the latter. This way, we simulate a realistic scenario where data have an __underlying regularity that we wish to learn__, but measurements are disturbed, say, by a random process.

```python
def config(axis):
    """
    Simple plot configuration for y(x) = sin(2 np.pi x)
    """
    axis.set(ylabel='y(x)',
             xlabel='x',
             ylim=(-2.4, 2.4),              # Limit range of y values
             xticks=np.linspace( 0, 1, 5),  # Set x- and y-tick marks
             yticks=np.linspace(-2, 2, 5))


# Create a random number generator (with given seed)
rng = np.random.RandomState(123456789)

# Generate 10 equally-spaced values in range 0 to 1
x = np.linspace(0, 1, 10)

# Compute y(x), adding noise from 
# a normal distribution N(0, 0.3)
y = np.sin(2 * np.pi * x) + rng.normal(scale=0.3, size=10)

# Expected y(x)
a = np.linspace(0, 1, 1000)
b = np.sin(2 * np.pi * a)

_, ax = plt.subplots()

# Configure the plot
config(ax)

ax.scatter(x, y, color='#3B5B72')
ax.plot(a, b, color='coral', linewidth=2, linestyle=":");
```

Our __task__ is to discover $y(x) = \sin(2\pi x)$ from a finite data set by __fitting the training data__ with a  polynomial function. 

Let's consider, for now, a polynomial of _degree_ (or _order_) 4:

\begin{equation}
y(x, \mathbf{w}) = w_1x + w_2x^2 + w_{3}x^{3} + w_{4}x^{4}
\end{equation}

The polynomial will be determined by the __coefficients__ $w_1, \dots, w_4$. How is it done?

The values of the coefficients $w_1, \dots, w_4$ are determined by minimising an __error function__ that measures misfit between $y(x, \mathbf{w})$ and the training data $\mathbf{x}$ and target values $\mathbf{t}$. 

A widely used error function is the sum-of-squares error $E$:

\begin{equation}
E(\mathbf{w}) = \frac{1}{2} \sum_{i=1}^{n} \lbrace y(x_i, \mathbf{w}) - t_i \rbrace^2
\end{equation}

where $n$ is the number of input samples.

Note that the function is $0$ if, and only if, $y(x_i, \mathbf{w})$ predicts exactly the target values $\mathbf{t}$. So, ideally, the __best solution__ is the one with the smallest error. The solution can be found in closed form:

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline

# A polynomial regression model of degree 4
model = make_pipeline(PolynomialFeatures(4), LinearRegression())
model.fit(x[:, np.newaxis], y)

# Predict y-value for 1,000 data points in range from 0 to 1
u = np.linspace(0, 1, 1000)
v = model.predict(u[:, np.newaxis])

_, ax = plt.subplots()

# Configure the plot
config(ax)

# Overlay the predicted curve over the input data
ax.scatter(x, y, color='#3B5B72')
ax.plot(a, b, color='coral', linewidth=2, linestyle=":")
ax.plot(u, v, color='#B7BF35', linewidth=3);
```

### Why $m = 4$ coefficients?

The remaining problem is to answer: __how many parameters?__ Let's try $m = 1$, $m = 2$, $m =4$ and $m = 10$.

```python
def _config(axis, title, i, j):
    """
    Simple plot configuration for y(x) = sin(2 np.pi x)
    for subplots
    """
    axis.set(title=title)
    # Since we are sharing x- and y-axis...
    if j < 1:
        axis.set(ylabel='y(x)')
    if i > 0:
        axis.set(xlabel='x')

    axis.set(ylim=(-2.4, 2.4),
             xticks=np.linspace( 0, 1, 5),
             yticks=np.linspace(-2, 2, 5))

    
figure, axes = plt.subplots(2, 2, 
                            sharex=True, 
                            sharey=True, 
                            constrained_layout=True, 
                            figsize=(10, 8))


# Predict y-value for 1,000 data points in range from 0 to 1
u = np.linspace(0, 1, 1000)

for m, i, j in zip([1, 2, 4, 10], [0, 0, 1, 1], [0, 1, 0, 1]):
    #
    # Create and train a polynomial regression model 
    # with m parameters
    #
    model = make_pipeline(PolynomialFeatures(m), LinearRegression())
    model.fit(x[:, np.newaxis], y)
    v = model.predict(u[:, np.newaxis])

    _config(axes[i, j], '$m = {}$'.format(m), i, j)
    axes[i, j].scatter(x, y, color='#3B5B72')
    axes[i, j].plot(a, b, color='coral', linewidth=2, linestyle=":")
    axes[i, j].plot(u, v, color='#B7BF35', linewidth=3)
```

We can make three obvservations from the results:

__1)__ The first- ($m=1$) and second-order ($m=2$) polynomials result in rather poor performance.

__2)__ The forth order polynomial ($m=4$) is a _good fit_ of the function $y(x) = \sin(2\pi x)$.

__3)__ A much higher order polynomial ($m = 10$) results in an excellent fit to the training data. In fact, it passes through each point. But, in reality, it is a poor approximation of $\sin(2\pi x)$.

The latter behaviour is known as __over-fitting__. Note that the __goal of machine learning__, on the other hand, is __good generalisation__; the ability to predict new data.

### What has happened when $m=10$?

We have 10 sample data points (the blue points in the graphs); and 10 parameters $w_1, \dots, w_{10}$. So, each parameter $w$ can be fine-tuned to match exactly each point (including noise).

But for points in the middle, the polynomial oscillates a lot: in order to fit each data point, the values of $w$ become too large!

So, we could __throw more data to the problem__ (100 instead of 10)...

```python
rng = np.random.RandomState(123456789)

x = np.linspace(0, 1, 100)
y = np.sin(2 * np.pi * x) + rng.normal(scale=0.3, size=100)

model = make_pipeline(PolynomialFeatures(10), LinearRegression())
model.fit(x[:, np.newaxis], y)

# Predict y-value for 1,000 data points in range from 0 to 10
u = np.linspace(0, 1, 1000)
v = model.predict(u[:, np.newaxis])

_, ax = plt.subplots()

# Configure the plot
config(ax)

# Overlay the predicted curve over the input data
ax.set(title="$m=10$")
ax.scatter(x, y, color='#3B5B72')
ax.plot(a, b, color='coral', linewidth=2, linestyle=":")
ax.plot(u, v, color='#B7BF35', linewidth=3);
```

We observe that the over-fitting problem becomes less severe with more training data. 

Roughly speaking, we can say that the size of the data set should be $5\times$ or $10\times$ the size of the model (i.e., the number of parameters). But this is not good...

In other words, we want to the size of the model to be related to the complexity of the problem solved, not the size of the dataset, right?

## Regularisation

One technique to address the over-fitting problem is __regularisation__; that is, introduce a penalty term to the error function so that values of $w$ do not get too large. The penalty, for example, can be the sum of $w_{i}^2$. The importance of the penalty is controlled by parameter $\alpha$:

\begin{equation}
E(\mathbf{w}) = \frac{1}{2} \sum_{i=1}^{n} \lbrace y(x_i, \mathbf{w}) - t_i \rbrace^2 + \frac{\alpha}{2} \sum_{j=1}^{m} {w_j}^2
\end{equation}

This particular case is known as __ridge regression__, but in general this technique in machine learning is known as _weight decay_.

If the penalty term is $\frac{\alpha}{2} \sum_{j=1}^{m} \vert {w_j} \vert$, i.e. we simply sum the parameters $w$ and not square them, it is known as __lasso regression__.

Note that now we have one more __hyper-parameter__ to control, $\alpha$.

### Ridge regression

```python
from sklearn.linear_model import Ridge

x = np.linspace(0, 1, 10)
# noise = rng.normal(scale=0.3, size=50)
y = np.sin(2 * np.pi * x) + rng.normal(scale=0.3, size=10)

model = make_pipeline(PolynomialFeatures(10), Ridge(alpha=0.0001)) # 0.0001))
model.fit(x[:, np.newaxis], y)

# Predict y-value for 1,000 data points in range from 0 to 1
u = np.linspace(0, 1, 1000)
v = model.predict(u[:, np.newaxis])

_, ax = plt.subplots()

# Configure the plot
config(ax)

# Overlay the predicted curve over the input data
ax.scatter(x, y, color='#3B5B72')
ax.plot(a, b, color='coral', linewidth=2, linestyle=":")
ax.plot(u, v, color='#B7BF35', linewidth=3);
```

### Lasso regression

```python
from sklearn.linear_model import Lasso

model = make_pipeline(PolynomialFeatures(10), Lasso(alpha=0.0001, max_iter=1000000))
model.fit(x[:, np.newaxis], y)

# Predict y-value for 1,000 data points in range from 0 to 10
u = np.linspace(0, 1, 1000)
v = model.predict(u[:, np.newaxis])

_, ax = plt.subplots()

# Configure the plot
config(ax)

# Overlay the predicted curve over the input data
ax.scatter(x, y, color='#3B5B72')
ax.plot(a, b, color='coral', linewidth=2, linestyle=":")
ax.plot(u, v, color='#B7BF35', linewidth=3);
```

### Why is regularisation important?

A regulariser is one of the methods to __shrink parameter values towards zero__. 

Thus, it is an opportunity to derive __sparse models__. No longer do we have to worry much about the size of the model with respect to the input data. The _effective number of parameters_ (that is, the non-zero ones) automatically adapt to the problem.



## Hyper-parameters

Let's recap the number of __hyper-parameters__ that we have to tune to solve a regression problem:

1. The choice of a basis function (e.g. $\phi_i(x) = x^i$)
2. The model size (i.e., the number of parameters $w$)
3. The parameter $\alpha$ of the regulariser
4. The number of iterations (depending on the solver)
