```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# load dataset
df = pd.read_csv("WHR2023.csv")
df.head(5)
```

```python
# correlation heatmap
plt.figure(figsize = (12, 8))
corr = df[["Life Ladder","Log GDP per capita", "Social support","Healthy life expectancy at birth", "Freedom to make life choices", 
    "Generosity", "Perceptions of corruption", "Positive affect","Negative affect"]].corr()
```

```python
corr
```

```python
# show correlation with heatmap
sns.heatmap(corr, annot = True, cmap = "coolwarm", fmt = '.2f', linewidth = 0.5)
plt.title("Correlation Heatmap of Happiness factors")
```

```python
# scatter plot

plt.figure(figsize = (10, 6))
afg = df[df["Country name"] == "Afghanistan"]
afg.head()

sns.scatterplot(x="year", y="Log GDP per capita", data=df, size="Social support", hue="Generosity",
                palette = "viridis", sizes=(20, 200))
plt.title("Year vs GDP")
plt.xlabel("Year")
plt.ylabel("GDP")
plt.show()
```

```python
plt.figure(figsize = (10, 6))
afg = df[df["Country name"] == "Afghanistan"]
afg.head()

sns.scatterplot(x="Life Ladder", y="Log GDP per capita", data=df, size="Social support", hue="Generosity",
                palette = "viridis", sizes=(20, 200))
plt.title("Life Ladder vs Log GDP per capita")
plt.xlabel("Life Ladder")
plt.ylabel("Log GDP per capita")
plt.show()
```

```python
plt.figure(figsize = (10, 6))

sns.boxplot(x="year", y="Life Ladder", data=df, palette = "viridis")
plt.title("Year vs Life Ladder")
plt.xlabel("Year")
plt.ylabel("Life Ladder")
plt.show()
```
