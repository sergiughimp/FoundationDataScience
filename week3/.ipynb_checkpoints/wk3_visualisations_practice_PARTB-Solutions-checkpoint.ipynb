# Introduction to Environmental Data Analysis

In this exercise, we will work with a dataset that records various environmental factors over time, such as temperature, pressure, wind speed, and particulate matter (PM2.5) levels. The goal of this exercise is to explore how these environmental variables behave, uncover trends, and detect anomalies that may reveal interesting insights into environmental patterns and especially to use the best visualization procedure to achieve this goal.

## Objectives

1. **Time Series Consistency**: We begin by ensuring that the time data is clean and consistent. This step is crucial for any time-based analysis, as missing, unsorted or duplicated timestamps can skew our results.
   
2. **Monthly Averages**: After verifying the data consistency, we will aggregate the data to calculate monthly averages for key variables. Monthly aggregation provides a clearer view of long-term trends by smoothing out short-term fluctuations.

3. **Data Visualization**: Visualizing data is essential for identifying trends and patterns that may not be immediately obvious from raw numbers. You will create the following plots to explore the behavior of temperature and its relationship with other variables:
    - **Line Plot**: A time series plot of monthly average temperatures to observe trends over time.
    - **Bar Chart**: A bar chart displaying monthly temperature anomalies, making it easy to compare temperature deviations across months.
    - **Histogram**: A histogram to visualize the distribution of monthly temperature anomalies.
    - **Scatter Plot**: A scatter plot showing the relationship between temperature and pressure, enhanced with a trendline to observe correlations.



4. **Anomaly analysis**: Finally, you will compute temperature anomalies with respect to the year 2010 as a baseline. Anomalies highlight deviations from the expected norm and can help detect unusual environmental changes, such as unusually hot or cold months.


## Dataset Information

This dataset provides weather and pollution data collected hourly over five years at the US embassy located in Beijing, China. The data includes information such as PM2.5 concentration (a measure of pollution) and various weather parameters, including dew point, temperature, pressure, wind direction, wind speed, and the cumulative number of hours of snow and rain.

### Data Preprocessing Steps

1. Load the raw dataset and use the time column as the DataFrame index.

2. Handle missing values. A quick check reveals scattered NA values for PM2.5 later in the dataset, which we will remove.

3. Retain only the column related to PM2.5 concentration, Dew Point, Temperature, Pressure, Cumulated wind speed,Cumulated hours of snow ,Cumulated hours of rain

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the dataset
file_path = 'Enviromental_dataset.csv'
data_new = pd.read_csv(file_path)


## Using the time column as the index



# Convert 'time' to datetime and set it as the index
data_new['time'] = pd.to_datetime(data_new['time'], format='%d/%m/%Y %H:%M')
data_new.set_index('time', inplace=True)
```

# Sort the dataset by time to ensure it's in chronological order


```python
data_new = data_new.sort_index()
```

## Drop the row containing NAs


```python
data_new.dropna(inplace=True)
data_new.head()
```

## Retain only selected columns


```python
data_new=data_new[['pm2.5','DEWP','TEMP','PRES','Iws','Is','Ir']]
data_new.head(3)
```

```python
# Check for any missing or duplicate datetime entries

missing_times_new = data_new.index.isnull().sum()
duplicate_times_new = data_new.index.duplicated().sum()


missing_times_new, duplicate_times_new
```

```python
# Create a monthly average dataframe
monthly_avg_data = data_new.resample('ME').mean()

# Create a new column for the month names
monthly_avg_data['month_name'] = monthly_avg_data.index.month_name()

# Display the first few rows of the monthly average data
monthly_avg_data[['TEMP', 'PRES']].head()
```

## Create the visualizations for temperature and pressure using the monthly average data


### Create a plot that shows the monthly average temperature over time

```python

plt.figure(figsize=(12, 8))

# 1. Line plot for monthly average temperature over time with trendline
#plt.subplot(2, 2, 1)
x_time = np.arange(len(monthly_avg_data.index))
y_temp = monthly_avg_data['TEMP']
trend_coeffs = np.polyfit(x_time, y_temp, 1)
trendline = np.poly1d(trend_coeffs)

plt.plot(monthly_avg_data.index, y_temp, label='Temperature', color='blue')
plt.plot(monthly_avg_data.index, trendline(x_time), label='Trendline', color='red', linestyle='--')
plt.title('Monthly Average Temperature Over Time with Trendline')
plt.xlabel('Time')
plt.ylabel('Temperature (°C)')
plt.xticks(rotation=45)
plt.legend()
```

### Create a plot that can represent the relationship between pressure and temperature


```python
# 2. Scatter plot of monthly average temperature vs pressure with trendline
#plt.subplot(1, 1, 1)
x_press = monthly_avg_data['PRES']
y_temp_press = monthly_avg_data['TEMP']

press_temp_trend_coeffs = np.polyfit(x_press, y_temp_press, 1)
press_temp_trendline = np.poly1d(press_temp_trend_coeffs)

plt.scatter(x_press, y_temp_press, alpha=0.5, color='purple')
plt.plot(x_press, press_temp_trendline(x_press), color='red', linestyle='--', label='Trendline')
plt.title('Temperature vs Pressure with Trendline')
plt.xlabel('Pressure (hPa)')
plt.ylabel('Temperature (°C)')
plt.legend()

plt.tight_layout()
plt.show()
```

### Create a chart for monthly average temperature by month


```python
# 3. Bar chart for monthly average temperature anomaly by month
monthly_avg_anomaly_by_month = monthly_avg_data.groupby('month_name')['TEMP'].mean()
monthly_avg_anomaly_by_month = monthly_avg_anomaly_by_month.reindex(['January', 'February', 'March', 'April', 'May', 'June',
                                                                     'July', 'August', 'September', 'October', 'November', 'December'])
```

```python
#plt.subplot(1, 1, 1)
monthly_avg_anomaly_by_month.plot(kind='bar', color='green')
plt.title('Average Temperature by Month')
plt.xlabel('Month')
plt.ylabel('Temperature (°C)')
plt.xticks(rotation=45)
plt.show()
```

### Create a plot showing the distribution of temperature


```python
# 4. Histogram of temperature anomalies
#plt.subplot(1, 1, 1)
plt.hist(monthly_avg_data['TEMP'], bins=10, color='orange', alpha=0.7)
plt.title('Histogram of Temperature')
plt.xlabel('Temperature (°C)')
plt.ylabel('Frequency')
```

### Analysis related to anomalies

```python
# Compute the 2010 average temperature
temp_2010_avg_new = monthly_avg_data[monthly_avg_data.index.year == 2010]['TEMP'].mean()

# Calculate anomalies: difference between actual monthly average temperature and the 2010 average
monthly_avg_data['TEMP_anomaly'] = monthly_avg_data['TEMP'] - temp_2010_avg_new

# Display the updated dataframe
monthly_avg_data.head()
```

## Do the same visualisations using tempeature anomalies

```python
# 1. Line plot for monthly average temperature over time with trendline
x_time = np.arange(len(monthly_avg_data.index))
y_temp = monthly_avg_data['TEMP_anomaly']
trend_coeffs = np.polyfit(x_time, y_temp, 1)
trendline = np.poly1d(trend_coeffs)

plt.plot(monthly_avg_data.index, y_temp, label='Temperature Anomalies', color='blue')
plt.plot(monthly_avg_data.index, trendline(x_time), label='Trendline', color='red', linestyle='--')
plt.title('Monthly Average Temperature Over Time with Trendline')
plt.xlabel('Time')
plt.ylabel('Temperature Anomalies (°C)')
plt.xticks(rotation=45)
plt.legend()
```

```python
# 2. Scatter plot of monthly average temperature vs pressure with trendline
x_press = monthly_avg_data['PRES']
y_temp_press = monthly_avg_data['TEMP_anomaly']

press_temp_trend_coeffs = np.polyfit(x_press, y_temp_press, 1)
press_temp_trendline = np.poly1d(press_temp_trend_coeffs)

plt.scatter(x_press, y_temp_press, alpha=0.5, color='purple')
plt.plot(x_press, press_temp_trendline(x_press), color='red', linestyle='--', label='Trendline')
plt.title('Temperature vs Pressure with Trendline')
plt.xlabel('Pressure (hPa)')
plt.ylabel('Temperature (°C)')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# 3. Bar chart for monthly average temperature anomaly by month
monthly_avg_anomaly_by_month = monthly_avg_data.groupby('month_name')['TEMP_anomaly'].mean()
monthly_avg_anomaly_by_month = monthly_avg_anomaly_by_month.reindex(['January', 'February', 'March', 'April', 'May', 'June',
                                                                     'July', 'August', 'September', 'October', 'November', 'December'])
monthly_avg_anomaly_by_month.plot(kind='bar', color='green')
plt.title('Average Temperature Anomaly by Month')
plt.xlabel('Month')
plt.ylabel('Temperature Anomaly (°C)')
plt.xticks(rotation=45)
plt.show()
```

```python
# 4. Histogram of temperature anomalies
plt.hist(monthly_avg_data['TEMP_anomaly'], bins=10, color='orange', alpha=0.7)
plt.title('Histogram of Temperature Anomalies')
plt.xlabel('Temperature Anomaly (°C)')
plt.ylabel('Frequency')

```
